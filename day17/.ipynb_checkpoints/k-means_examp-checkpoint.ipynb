# K-Means Clustering

Clustering is a type of unsupervised learning. This is very often used when you don't have labeled data. K-Means Clustering is one of the popular clustering algorithm. The goal of this algorithm is to find the groups(clusters) in the given data.

K-Means is a very simple algorithm which clusters the data into K number of clusters.

# Use Cases

K-Means is widely used for many applications.

1. Image segmentation
2. Clustering Gene Segmentation Data
3. News Article Clustering
4. Clustering languages
5. Species Clustering
6. Anomaly Detection

# Algorithm

```python
%matplotlib inline
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (16,9)
plt.style.use('ggplot')
```

```python
# Importing the dataset
data = pd.read_csv('xclara.csv')
print(data.shape)
data.head()
```

```python
# Getting the values and plotting it
f1 = data['V1'].values
f2 = data['V2'].values
#print (zip(f1,f2))
X = np.array(list(zip(f1,f2)))
plt.scatter(f1,f2,c='black',s=7) # c is color and s is size
```

```python
# Euclidean Distance Calculator
def dist(a,b,ax=1):
    return np.linalg.norm(a - b, axis=ax)
```

```python
# Number of clusters
k=3
# X coordinates of random centroids
C_x = np.random.randint(0,np.max(X)-20, size=k)
# Y coordinates of random centroids
C_y = np.random.randint(0,np.max(X)-20, size=k)
C = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print(C)
```

```python
#plotting along with the Centroids
plt.scatter(f1,f2,c='#050505', s=7)
plt.scatter(C_x,C_y,marker='*',s=200,c='g')
```

```python
# To store the value of centroid when it updates
C_old = np.zeros(C.shape)
# Cluster Lables(0,1,2)
clusters = np.zeros(len(X)) #len is 3000
# Error func. - Distance between new centroids and old centroids
error = dist(C, C_old, None)
# Loop will run till the error becomes zero
while error != 0:
    # Assigning each value to its closest cluster
    for i in range(len(X)):
        distances = dist(X[i],C)
        cluster = np.argmin(distances)
        clusters[i] = cluster
    # Storing the old centroid values
    C_old = deepcopy(C)
    # Finding the new centroids by taking the average value
    for i in range(k):
        points = [X[j] for j in range(len(X)) if clusters[j] == i]
        C[i] = np.mean(points, axis=0)
    error = dist(C, C_old, None)
```

```python
colors = ['r','g','b','y','c','m']
fig, ax = plt.subplots()
for i in range(k):
    points = np.array([X[j] for j in range(len(X)) if clusters[j] == i])
    ax.scatter(points[:, 0], points[:, 1], s=7, c=colors[i])
ax.scatter(C[:, 0], C[:, 1], marker='*', s=200, c='#050505')
```

# The scikit-learn approach

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)
# Getting the cluster Labels
labels = kmeans.predict(X)
print(labels) #kaunsa kis cluster me h (0,1,2) as teen h
# Centroid values
centroids = kmeans.cluster_centers_
```

```python
# Comparing with scikit-learn centroids
print(C) # From Scratch
print(centroids) # From sci-kit learn
```

we will generate a new dataset using make_blobs function.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize'] = (16,9)

# Creating a sample dataset with 4 clusters
X,y = make_blobs(n_samples=800, n_features=3, centers=4)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

```python
# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```
